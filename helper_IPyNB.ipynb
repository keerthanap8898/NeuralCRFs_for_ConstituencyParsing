```python

#file1 = open('outputs/2000/output.txt', 'r')
file1 = open('outputs/full/output.txt', 'r')
#file1 = open('outputs/lt/output.txt', 'r')
Lines = file1.readlines()
uf, lf, x=[], [], []
c=0
for line in Lines:
    if "dev:  loss:" in line:
        line=line.split()
        print(line)
        c+=1
        x.append(c)
        uf.append(float(line[13].replace('%','')))
        lf.append(float(line[19].replace('%','')))
print(uf)
print(lf)
print(x)   
        #loss 2, ucm 5, lcm 7, up 9, ur 11, uf 13, lp 15, lr 17, lf 19
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x, uf, color='red')
plt.plot(x, lf, color='green')
plt.title("UF-LF loss graph (uf-red, lf-green)")
plt.ylabel("UF/LF")
plt.xlabel("epochs")    
plt.show()
#plt.savefig('UF_LF.png')
```
